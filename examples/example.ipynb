```python
list_of_keys = F.keys()
print('{} ...'.format(list_of_keys[:5]))
```

These keys can be indexed a string or list of string

```python
S = F['135.1']

S_multple = F[['424.1', '426.1']]
```

```python
S_range = F[424:426]
```

automatic plot assuming you want to plot the last index against the first. Clearly the last column is not so ineresting in this file!

```python
_ = S.plot()
```

```python
_ = S.plot(xkey='Epoch', ykey='mycntr3', monitor='Seconds')
```
